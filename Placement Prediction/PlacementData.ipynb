```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
df['ExtracurricularActivities'] = label_encoder.fit_transform(df['ExtracurricularActivities'])
df['PlacementTraining'] = label_encoder.fit_transform(df['PlacementTraining'])
df['PlacementStatus'] = label_encoder.fit_transform(df['PlacementStatus'])
```

```python
#basically we have encoded the dataframe and 0 means not placed and 1 means placed
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('PlacementStatus' , axis = 1)
y = df['PlacementStatus']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=1000)
```

```python
rf.fit(X_train,y_train)
```

```python
pred = rf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
pred1 = dtree.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred1))
print(classification_report(y_test,pred1))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
pred2 = log.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred2))
print(classification_report(y_test,pred2))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error_rate = []

for i in range(1,40):
    KNN = KNeighborsClassifier(n_neighbors=i)
    KNN.fit(X_train,y_train)
    pred3 = KNN.predict(X_test)
    error_rate.append(np.mean(pred3 != y_test))
```

```python
sns.set_style('darkgrid')
figure = plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,ls='--',marker = 'o' , markerfacecolor = 'red' ,markersize = 10)
plt.title('K value vs Error Rate')
plt.xlabel('K Value')
plt.ylabel('Error Rate')
```

```python
#We see that the error rate is least when n_neighbours is 11
KNN = KNeighborsClassifier(n_neighbors=11)
KNN.fit(X_train,y_train)
pred4 = KNN.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred4))
print(classification_report(y_test,pred4))
```
